```python
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
import numpy as np
from itertools import product
```

```python
data = [[0, 1], [0, 0], [1, 0], [1, 1]] # AND and XOR gate dataset

labels = [0, 0, 0, 1] #AND gate labels

x = [point[0] for point in data] 
y = [point[1] for point in data]

plt.scatter(x, y, c=labels)
plt.show()
```

```python
classifier = Perceptron(max_iter= 40) #initialize the perceptron classifier
classifier.fit(data, labels) #fit the perceptron classifier
classifier.score(data, labels) #check the performance
```

Visualizing the Perceptron

```python
x_values = np.linspace(0, 1, 100) #create random numbers 0 and 100
y_values = np.linspace(0, 1, 100)
point_grid = list(product(x_values, y_values))

distances = classifier.decision_function(point_grid) #determine the decision boundary 

abs_distances = [abs(i) for i in distances]

distances_matrix = np.reshape(abs_distances, (100, 100)) #create 100 x values and 100 y values

heatmap = plt.pcolormesh(x_values, y_values, distances_matrix) #plot a heatmap 
plt.colorbar(heatmap)

plt.scatter(x, y, c=labels)
plt.show()
```
